```python
df = pd.DataFrame(run_history_all)
df.head()
```

```python
df = df[df.name != "VICReg ImageNet LP HLR - chexpert"]
df.head(35)
```

```python
def extract_method(x):
    return x.split(' ')[0]

def extract_base_ds(x):
    return x.split(' ')[1]

def extract_tune(x):
    return x.split(' ')[2]

def extract_task(x):
    return x.split(' ')[4]
```

```python
df['method'] = df.name.apply(extract_method)
df['base_ds'] = df.name.apply(extract_base_ds)
df['tune'] = df.name.apply(extract_tune)
df['task'] = df.name.apply(extract_task)
```

```python
df.sort_values(by=['task','tune'])
```

```python
df.to_csv("results.csv")
```
