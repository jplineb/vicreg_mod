# Look at valid distribution

```python
path = "/zfs/wficai/chexpert/chexpertchestxrays-u20210408/valid.csv"
df = pd.read_csv(path)
df.head()
```

```python
len(df)
```

```python
df['Fracture'].value_counts()
```

```python
df["AP/PA"].value_counts()
```

```python
df[df['AP/PA'].isin(["PA", "AP"])].head()
```

```python
# Check distribution of labels 
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=len(df.columns[5:]), ncols=1, figsize=(5,45))
fig.suptitle("Valid PA View Label Distributions")
fig.tight_layout(pad=5.0)
# Check distribution of labels 
i = 0
for column in df.columns[5:]:
    print(column)
    print(df[df['AP/PA'] == "PA"][column].value_counts(dropna=False).plot(kind="bar", ax=axes[i], title = column))
    i +=1
```

```python
# Check how much data we used for valid in PA View
print(df['AP/PA'].value_counts(dropna=False))
print(f"fraction of total used for validation: {df['AP/PA'].value_counts(dropna=False)['PA']/len(df)}")
```
